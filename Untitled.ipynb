```python
def base_to_index(base):
    return {'A': 0, 'C': 1, 'G': 2, 'T': 3}[base]

def sample_r_prior(MAX_STATES):
    return max(1, int(torch.poisson(torch.tensor([3.0], device=device)).item()) % (MAX_STATES + 1))

def sample_q_prior(MAX_ORDER):
    return int(torch.poisson(torch.tensor([1.0], device=device)).item()) % (MAX_ORDER + 1)

def sequence_to_indices(sequence):
    return torch.tensor([base_to_index(b) for b in sequence], dtype=torch.long, device=device)

def initialize_parameters(n_states, n_bases, q):
    n_states = int(n_states)  # Ensure n_states is an integer
    n_bases = int(n_bases)    # Ensure n_bases is an integer
    Lambda = Dirichlet(torch.ones(n_states, device=device)).sample((n_states,)).to(device)
    P = Dirichlet(torch.ones(int(n_bases**(q+1)), device=device)).sample((n_states,)).to(device)
    return Lambda, P

def calculate_posterior_probs(all_forward, all_backward):
    all_posterior_probs = []
    for F, B in zip(all_forward, all_backward):
        posterior_probs = F * B
        posterior_probs /= torch.sum(posterior_probs, dim=0)
        all_posterior_probs.append(posterior_probs)
    return all_posterior_probs

def update_q_multi(sequences, r, q, Lambda, P, MAX_ORDER):
    log_probs = torch.zeros(MAX_ORDER + 1, device=device)
    for new_q in range(MAX_ORDER + 1):
        new_P = Dirichlet(torch.ones(int(4**(new_q+1)), device=device)).sample((r,)).to(device)
        log_probs[new_q], _, _ = forward_multi(sequences, Lambda.to(device), new_P.to(device), new_q)
    
    new_q = torch.multinomial(torch.exp(log_probs - torch.logsumexp(log_probs, dim=0)), 1).item()
    if new_q != q:
        P = Dirichlet(torch.ones(int(4**(new_q+1)), device=device)).sample((r,)).to(device)
    
    return new_q, P

def rjmcmc_r_multi(sequences, r, q, Lambda, P, MAX_STATES):
    Lambda = Lambda.to(device)
    P = P.to(device)
    if torch.rand(1, device=device).item() < 0.5:  # Attempt birth
        if r < MAX_STATES:
            new_r = r + 1
            new_Lambda = torch.zeros((new_r, new_r), device=device)
            new_Lambda[:r, :r] = Lambda
            new_Lambda[r, :] = Dirichlet(torch.ones(new_r, device=device)).sample().to(device)
            new_Lambda[:, r] = Dirichlet(torch.ones(new_r, device=device)).sample().to(device)
            new_P = torch.vstack((P, Dirichlet(torch.ones(int(4**(q+1)), device=device)).sample().to(device)))
            
            # Calculate acceptance probability
            log_A, _, _ = forward_multi(sequences, new_Lambda, new_P, q)
            log_A -= forward_multi(sequences, Lambda, P, q)[0]
            log_A += torch.log(torch.tensor(new_r, device=device)) - torch.log(torch.tensor(r + 1, device=device))  # Prior ratio and proposal ratio
            
            if torch.log(torch.rand(1, device=device)).item() < log_A:
                return new_r, new_Lambda, new_P
    else:  # Attempt death
        if r > 1:
            new_r = r - 1
            j = torch.randint(r, (1,), device=device).item()
            new_Lambda = torch.cat([torch.cat([Lambda[:j, :j], Lambda[:j, j+1:]], dim=1),
                                    torch.cat([Lambda[j+1:, :j], Lambda[j+1:, j+1:]], dim=1)], dim=0)
            new_P = torch.cat([P[:j], P[j+1:]], dim=0)
            
            # Calculate acceptance probability
            log_A, _, _ = forward_multi(sequences, new_Lambda, new_P, q)
            log_A -= forward_multi(sequences, Lambda, P, q)[0]
            log_A += torch.log(torch.tensor(r, device=device)) - torch.log(torch.tensor(new_r + 1, device=device))  # Prior ratio and proposal ratio
            
            if torch.log(torch.rand(1, device=device)).item() < log_A:
                return new_r, new_Lambda, new_P
    
    return r, Lambda, P

def run_mcmc_multi(sequences, n_iter, n_burn_in, MAX_STATES, MAX_ORDER, thin=1):
    r = sample_r_prior(MAX_STATES)
    q = sample_q_prior(MAX_ORDER)
    Lambda, P = initialize_parameters(r, 4, q)
    
    samples = []
    for i in tqdm(range(n_iter + n_burn_in)):
        # Update q and P
        q, P = update_q_multi(sequences, r, q, Lambda, P, MAX_ORDER)
        
        # Update Lambda
        for j in range(r):
            Lambda[j, :] = Dirichlet(Lambda[j, :] + 1).sample().to(device)
        
        # Reversible jump for r
        r, Lambda, P = rjmcmc_r_multi(sequences, r, q, Lambda, P, MAX_STATES)
        
        # Ensure Lambda and P are on the correct device after rjmcmc
        Lambda = Lambda.to(device)
        P = P.to(device)
        
        # Store samples after burn-in period and apply thinning
        if i >= n_burn_in and (i - n_burn_in) % thin == 0:
            log_likelihood, all_forward, all_scale = forward_multi(sequences, Lambda, P, q)
            all_backward = backward_multi(sequences, Lambda, P, q, all_scale)
            all_posterior_probs = calculate_posterior_probs(all_forward, all_backward)
            viterbi_paths = viterbi_multi(sequences, Lambda, P, q)
            samples.append((r, q, Lambda.clone(), P.clone(), all_posterior_probs, viterbi_paths))
    
    return samples


def ravel_multi_index(multi_index, dims):
    strides = torch.cumprod(torch.tensor(dims[:0:-1], dtype=torch.long, device=device), dim=0).flip(0)
    strides = torch.cat([strides, torch.tensor([1], device=device)])
    return sum(torch.tensor(i, device=device) * s for i, s in zip(multi_index, strides))

def forward_multi(sequences, Lambda, P, q):
    n_states, n_emissions = P.shape
    total_log_likelihood = 0
    all_forward = []
    all_scale = []
    
    for sequence in sequences:
        n_obs = len(sequence)
        F = torch.zeros((n_states, n_obs), dtype=torch.float32, device=device)
        scale = torch.zeros(n_obs, dtype=torch.float32, device=device)
        
        # Initialize
        initial_state = tuple(sequence[:q].cpu().numpy())
        emission_index = ravel_multi_index(initial_state + (sequence[q].item(),), (4,)*(q+1))
        F[:, q-1] = P[:, emission_index] / n_states
        scale[q-1] = torch.sum(F[:, q-1])
        F[:, q-1] /= scale[q-1]
        
        # Recurse
        for t in range(q, n_obs):
            state = tuple(sequence[t-q:t+1].cpu().numpy())
            emission_index = ravel_multi_index(state, (4,)*(q+1))
            F[:, t] = P[:, emission_index] * torch.sum(F[:, t-1].unsqueeze(1) * Lambda, dim=0)
            scale[t] = torch.sum(F[:, t])
            F[:, t] /= scale[t]
        
        total_log_likelihood += torch.sum(torch.log(scale)).item()
        all_forward.append(F)
        all_scale.append(scale)
    
    return total_log_likelihood, all_forward, all_scale

def backward_multi(sequences, Lambda, P, q, all_scale):
    Lambda = Lambda.to(device)
    P = P.to(device)
    n_states, n_emissions = P.shape
    all_backward = []
    
    for sequence, scale in zip(sequences, all_scale):
        sequence = sequence.to(device)
        scale = scale.to(device)
        n_obs = len(sequence)
        B = torch.zeros((n_states, n_obs), dtype=torch.float32, device=device)
        
        # Initialize
        B[:, -1] = 1 / scale[-1]
        
        # Recurse
        for t in range(n_obs-2, q-2, -1):
            state = tuple(sequence[t-q+1:t+2].cpu().numpy())
            emission_index = ravel_multi_index(state, (4,)*(q+1))
            B[:, t] = torch.sum(B[:, t+1].unsqueeze(0) * Lambda * P[:, emission_index], dim=1)
            B[:, t] /= scale[t]
        
        all_backward.append(B)
    
    return all_backward

def viterbi_multi(sequences, Lambda, P, q):
    Lambda = Lambda.to(device)
    P = P.to(device)
    n_states, n_emissions = P.shape
    segmentations = []
    
    for sequence in sequences:
        sequence = sequence.to(device)
        n_obs = len(sequence)
        V = torch.zeros((n_states, n_obs), dtype=torch.float32, device=device)
        path = torch.zeros((n_states, n_obs), dtype=torch.int32, device=device)
        
        # Initialize
        initial_state = tuple(sequence[:q].cpu().numpy())
        emission_index = ravel_multi_index(initial_state + (sequence[q].item(),), (4,)*(q+1))
        V[:, q-1] = torch.log(P[:, emission_index] / n_states)
        
        # Recurse
        for t in range(q, n_obs):
            state = tuple(sequence[t-q:t+1].cpu().numpy())
            emission_index = ravel_multi_index(state, (4,)*(q+1))
            prob = V[:, t-1].unsqueeze(1) + torch.log(Lambda) + torch.log(P[:, emission_index])
            V[:, t], path[:, t] = torch.max(prob, dim=0)
        
        # Backtrack
        best_path = torch.zeros(n_obs, dtype=torch.int32, device=device)
        best_path[-1] = torch.argmax(V[:, -1])
        for t in range(n_obs-2, q-2, -1):
            best_path[t] = path[best_path[t+1], t+1]
        
        segmentations.append(best_path)
    
    return segmentations
```
